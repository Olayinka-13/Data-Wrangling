## 1. Changing column names to lowercase

```python
df.columns=df.columns.str.lower()
```

## 2. Removing special characters from the column names and possibly renaming of columns

```python
df.rename(columns={'unnamed: 0': 'id', 'imbd title id': 'title id', 'original titlê': 'title', 'genrë¨': 'genre',
                   'unnamed: 8': 'unnamed',' votes ': 'votes'}, inplace=True)
```

## 3. Deleting the unnecessary columns

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df = df.drop(columns=['unnamed'])
```

```python
df.head()
```

## 4. Splitting the title id to id with numeric only

```python
df['title id']=df['title id'].str.split('tt', expand=True)[1]
```

```python
df.head()
```

## 5. Formating release year to show the year of release

```python
df['release year'].head(10)
```

```python
# Extract the column containing the release dates and assign it to a new variable
# date_column = df['release year']
```

```python
# Remove spaces in front of a number
df['release year']=df['release year'].str.strip()
```

```python
df['release year'] = df['release year'].str.replace('Jan', '01')
df['release year'] = df['release year'].str.replace('Feb', '02')
df['release year'] = df['release year'].str.replace('Mar', '03')
df['release year'] = df['release year'].str.replace('Apr', '04')
df['release year'] = df['release year'].str.replace('May', '05')
df['release year'] = df['release year'].str.replace('Jun', '06')
df['release year'] = df['release year'].str.replace('Jul', '07')
df['release year'] = df['release year'].str.replace('Aug', '08')
df['release year'] = df['release year'].str.replace('Sep', '09')
df['release year'] = df['release year'].str.replace('Oct', '10')
df['release year'] = df['release year'].str.replace('Nov', '11')
df['release year'] = df['release year'].str.replace('Dec', '12')
```

```python
# Replace separators
df['release year'] = df['release year'].str.replace(' ', '-')
df['release year'] = df['release year'].str.replace('--', '-')
df['release year'] = df['release year'].str.replace('/', '-')
```

```python
df['release year'].head(10)
```

```python
df.loc[1,'release year']='1972-09-21'
df.loc[2,'release year']='2008-07-23'
```

```python
df['release year']=pd.to_datetime(df['release year'], errors='coerce', format='%Y-%m-%d')
```

## 6. Selecting the genre that appeard first

```python
df.genre=df.genre.str.split(',', expand=True)[0]
```

```python
df
```

## 7. Standardized the country column to one prefix

```python
#value count
df.country.value_counts()
```

```python
#stripping space
df.country=df.country.str.strip()
```

```python
df['country'] = df['country'].replace(['US', 'US.'], 'USA')
df['country'] = df['country'].replace('West Germany', 'Germany')
df['country'] = df['country'].replace(['New Zesland', 'New Zeland'], 'New Zealand')
df['country'] = df['country'].replace('Italy1', 'Italy')
```

```python
df.country.value_counts()
```

## 8. Remove the dollar sign, change letter to number in income, take out commas, drop Null values and convert income to an interger

```python
df.income = df.income.str.split('$', expand=True)[1]
```

```python
df.head(5)
```

```python
df.income = df.income.str.replace('o', '0')
df.income = df.income.str.replace(",", "")
```

```python
df.head(5)
```

```python
#REPLACE nan with 0
df['income'] = df['income'].replace('NAN', 0)
#chabging the data type
df['income'] = df['income'].fillna(0).astype(np.int64)
```

```python
df.info()
```

## 9. Removing the . in the votes column and changing the data type

```python
df['votes'] = df['votes'].str.replace('.', '')
```

```python
df
```

```python
df['votes'] = df['votes'].fillna(0).astype(np.int64)
```

```python
df.info()
```

## 10. Changing duration to float and getting rid of non-numeric values

```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce', downcast='integer')
```

```python
df.info()
```

## 11. Modfiying the score column to standard format and data type

```python
df['score'] = df['score'].str.replace(",", "")
df['score'] = df['score'].str.replace("f", "")
df['score'] = df['score'].str.replace("e-0", "")
df['score'] = df['score'].str.replace("89", "8.9")
df['score'] = df['score'].str.replace("86", "8.6")
df['score'] = df['score'].str.replace("8..8", "8.8", regex=True)
df['score'] = df['score'].str.replace("8:8", "8.8", regex=True)
df['score'] = df['score'].str.replace("\+\+8\.7", "8.7", regex=True)
df['score'] = df['score'].str.replace("8.7.", "8.7", regex=True)
```

```python
df
```

```python
df['score'] = df['score'].astype(float)
```
